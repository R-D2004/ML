```python
# ---------------------------------------------------------
# Association Rule Learning using Apriori
# Dataset: Market Basket Optimization
# ---------------------------------------------------------

import numpy as np
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules
import matplotlib.pyplot as plt

# ---------------------------------------------------------
# 1. Load Dataset and Pre-processing
# ---------------------------------------------------------

# Dataset may be .csv or .xlsx depending on your file
# Example file name: "Market_Basket_Optimisation.csv"
dataset = pd.read_csv('Navy Bays_SMS SpamCollection.CSV', sep='\t', header=None)

# Assuming the message content is in the second column (index 1)
messages = dataset.iloc[:, 1].tolist()

# Convert messages into list of transactions (words in each message)
transactions = [message.split() for message in messages]

print("Sample transactions:")
print(transactions[:5])

# Convert list of transactions into One-Hot Encoded DataFrame
from mlxtend.preprocessing import TransactionEncoder

te = TransactionEncoder()
te_data = te.fit(transactions).transform(transactions)
df = pd.DataFrame(te_data, columns=te.columns_)

print("\nOne-Hot Encoded DataFrame:")
print(df.head())

# ---------------------------------------------------------
# 2. Apply Apriori Algorithm
# ---------------------------------------------------------

# Minimum support can be changed for experimentation
frequent_itemsets = apriori(df, min_support=0.02, use_colnames=True)

print("\nFrequent Itemsets:")
print(frequent_itemsets.head())

# ---------------------------------------------------------
# 3. Generate Association Rules
# ---------------------------------------------------------

rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.3)

print("\nAssociation Rules:")
print(rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']].head())

# ---------------------------------------------------------
# 4. Visualize Top Rules by Lift
# ---------------------------------------------------------

# Sort by highest lift
rules_sorted = rules.sort_values(by='lift', ascending=False).head(10)
print("\nTop 10 Rules Sorted by Lift:")
print(rules_sorted)

# Bar graph
plt.figure(figsize=(10,5))
plt.barh([str(i) for i in rules_sorted['antecedents']], rules_sorted['lift'])
plt.xlabel("Lift")
plt.ylabel("Antecedent Itemsets")
plt.title("Top 10 Association Rules by Lift")
plt.show()

# ---------------------------------------------------------
# 5. Change Support & Confidence Values to Observe Differences
# ---------------------------------------------------------

rules2 = association_rules(
    frequent_itemsets,
    metric="confidence",
    min_threshold=0.6  # Higher confidence
)
print("\nRules with confidence >= 0.6:")
print(rules2[['antecedents', 'consequents', 'support', 'confidence', 'lift']])
```
